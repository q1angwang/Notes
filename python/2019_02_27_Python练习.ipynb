### <font color="0000ff">4. 阶乘计算。计算$1!+2!+3!+\dots+10!$的结果。</font> 




```python
a = 1
su = 0
for j in range(1,10+1):
    for i in range(1, j+1):
        a = a*i
    su += a
    a = 1
print(su)
```
